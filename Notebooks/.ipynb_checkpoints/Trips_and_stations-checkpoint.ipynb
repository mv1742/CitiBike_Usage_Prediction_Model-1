```python
for i in range(24):
    print(i)
#     filter_sign = ' ' + str(i) + ':'
#     print(filter_sign)
#     tripHour_1 = tripdata[tripdata.apply(lambda x: x['starttime'][8:11] == filter_sign, axis=1)]
#     print(tripHour_1.shape)
    
#     tripHour_2 = tripdata[tripdata.apply(lambda x: x['starttime'][9:12] == filter_sign, axis=1)]
#     print(tripHour_2.shape)
    
#     tripHour_3 = tripdata[tripdata.apply(lambda x: x['starttime'][10:13] == filter_sign, axis=1)]
#     print(tripHour_3.shape)
    
#     tripHour = pd.concat([tripHour_1, tripHour_2])
#     tripHour = pd.concat([tripHour, tripHour_3])

#     print(tripHour.shape)

    tripstartHour = tripdata[tripdata.apply(lambda x: x['starttime'].hour == i, axis=1)]
    print(tripstartHour.shape)
    
    tripstartCount = tripstartHour.groupby('start station id').count()
    tripstartCount = tripstartCount.reset_index()[[0,1]]
    
    tripendHour = tripdata[tripdata.apply(lambda x: x['stoptime'].hour == i, axis=1)]
    print(tripendHour.shape)
    
    tripendCount = tripendHour.groupby('end station id').count()
    tripendCount = tripendCount.reset_index()[[0,2]]

    stations = stations.merge(tripstartCount, how='left', left_on='station_id', right_on='start station id')
    stations = stations.merge(tripendCount, how='left', left_on='station_id', right_on='end station id')
    
    stations.drop(['start station id', 'end station id'], axis=1, inplace=True)
    start_str = 'startCount_' + str(i)
    end_str = 'endCount_' + str(i)
    stations = stations.rename(index=str, columns={"tripduration": start_str, "starttime": end_str})
    
```
