```python
from __future__ import print_function
import pylab as pl
import os
import json
import sys
import time
import csv
import sched
try:
    import urllib2 as urllib
except ImportError:
    import urllib.request as urllib
```

```python
def read_real_time_data():
    url = "https://gbfs.citibikenyc.com/gbfs/en/station_status.json"

    # get data through api
    response = urllib.urlopen(url)
    data = response.read().decode("utf-8")
    data = json.loads(data)

    for item in data['data']['stations']:
        is_installed = item['is_installed']
        is_renting = item['is_renting']
        is_returning = item['is_returning']
        last_reported = item['last_reported']
        num_bikes_available = item['num_bikes_available']
        num_bikes_disabled = item['num_bikes_disabled']
        num_docks_available = item['num_docks_available']
        num_docks_disabled = item['num_docks_disabled']
        num_ebikes_available = item['num_ebikes_available']
        station_id = item['station_id']

        eightd_has_available_keys = item['eightd_has_available_keys']
        if 'eightd_active_station_services' in item.keys():
            id_num = item['eightd_active_station_services'][0]['id']
        else:
            id_num = None
```

```python
with open('station_status.csv',"w") as f:
    writer = csv.writer(f)
    writer.writerow(["is_installed","is_renting","is_returning","last_reported","num_bikes_available","num_bikes_disabled",
                     "num_docks_available","num_docks_disabled","num_ebikes_available","station_id","eightd_has_available_keys",
                    "id_num"])
    
    while True:
        url = "https://gbfs.citibikenyc.com/gbfs/en/station_status.json"

        # get data through api
        response = urllib.urlopen(url)
        data = response.read().decode("utf-8")
        data = json.loads(data)

        for item in data['data']['stations']:
            is_installed = item['is_installed']
            is_renting = item['is_renting']
            is_returning = item['is_returning']
            last_reported = item['last_reported']
            num_bikes_available = item['num_bikes_available']
            num_bikes_disabled = item['num_bikes_disabled']
            num_docks_available = item['num_docks_available']
            num_docks_disabled = item['num_docks_disabled']
            num_ebikes_available = item['num_ebikes_available']
            station_id = item['station_id']

            eightd_has_available_keys = item['eightd_has_available_keys']
            if 'eightd_active_station_services' in item.keys():
                id_num = item['eightd_active_station_services'][0]['id']
            else:
                id_num = None
                
            writer.writerow([is_installed,is_renting,is_returning,last_reported,num_bikes_available,num_bikes_disabled,
                             num_docks_available,num_docks_disabled,num_ebikes_available,station_id,eightd_has_available_keys,
                             id_num])
               
        time.sleep(600)
        print(1)
        
```
